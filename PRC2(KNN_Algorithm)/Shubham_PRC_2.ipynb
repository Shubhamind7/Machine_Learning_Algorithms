## Aim : Implement the Knn model for given datasets, obtain & plot the the train & test accuracy scores/values to graph for different. k values k=[1,3,5,7] 

#### Importing necessary Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

### [1] Iries Dataset

```python
dataset1 = pd.read_csv('iris.csv')
```

```python
dataset1.head(10)
```

```python
dataset1.info()
```

```python
X = dataset1.iloc[:, :-1].values
y = dataset1.iloc[:, 4].values
```

```python
for i in range(5):
    print(X[i])
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
no_neighbors = np.array([1,3,5,7])
train_accuracy1 = np.empty(len(no_neighbors))
test_accuracy1 = np.empty(len(no_neighbors))
print("Iries_Data_Modeling_KNN")
for i, k in enumerate(no_neighbors):
    # We instantiate the classifier
    knn = KNeighborsClassifier(n_neighbors=k)
    # Fit the classifier to the training data
    knn.fit(X_train,y_train)
    
    # Compute accuracy on the training set
    train_accuracy1[i] = knn.score(X_train, y_train)

    # Compute accuracy on the testing set
    test_accuracy1[i] = knn.score(X_test, y_test)
    
    #for k value
    print("K =",k)
    print("y_pred values:")
    y_pred = knn.predict(X_test)
    print(y_pred)
    print("confusion_matrix :")
    print(classification_report(y_test, y_pred))
    print("Final_Report:")
    print("Training_Score:",knn.score(X_train,y_train))
    print("Testing_Score:",knn.score(X_test,y_test))
    print("\n")
```
