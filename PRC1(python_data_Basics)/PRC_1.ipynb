### [3] Class in Python

```python
class BankAccount(object):
    def __init__(self,initial_balance=0):
        self.balance=initial_balance
    def deposit(self,amount):
        self.balance +=amount
    def withdrawal(self,amount):
        self.balance -=amount
    def overdrawn(self):
        return self.balance < 0
my_account=BankAccount(1000)
withdrawal=input("Amount you Want to take: ")
my_account.withdrawal(int(withdrawal))
print(my_account.balance,my_account.overdrawn())
```

### [6] Matpotlib 

####   (A) Sine_Cosine

```python
#import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0,3*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

#Plotting
plt.plot(x,y_sin)
plt.plot(x,y_cos)
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
plt.title('SIne & Cosine')
plt.legend('sine','cosine')
plt.show
```

####   (B) Line_Plot

```python
# x-axis values
x = [5, 2, 9, 4, 7]

# Y-axis values
y = [10, 5, 8, 4, 2]

#Labeling
plt.xlabel("X_axis")
plt.ylabel("Y_axis")

# Function to plot
plt.plot(x, y)

# function to show the plot
plt.show()
```

#### (C) BarPlot

```python
# x-axis values
x = [5, 2, 9, 4, 7]

# Y-axis values
y = [10, 5, 8, 4, 2]

# Function to plot
plt.bar(x, y)

#Labeling
plt.xlabel("X_axis Points")
plt.ylabel("Y_axis Points")

# function to show the plot
plt.show()
```
